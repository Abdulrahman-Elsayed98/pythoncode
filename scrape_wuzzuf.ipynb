```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page = requests.get(f'https://wuzzuf.net/search/jobs/?a=hpb%7Cspbg&q=data%20analyst')
soup = BeautifulSoup(page.content,'html')
soup
```

```python
titles = soup.find_all('a', attrs={"class": "css-o171kl", "rel": "noreferrer"})
names = [title.text for title in titles]
names
```

```python
organizations = soup.find_all('a', attrs={"class", "css-17s97q8"})
companies = [organization.text[:-1].strip() for organization in organizations]
companies
```

```python
places = soup.find_all('span', attrs={"class": "css-5wys0k"})
locations = [place.text.strip() for place in places]
locations
```

```python
times = soup.select("div.css-d7j1kk div")
date_posted = [time.text for time in times]
date_posted
```

```python
j_types = soup.find_all('span', attrs={"class": "css-1ve4b75 eoyjyou0"})
job_type = [j_type.text for j_type in j_types]
job_type
```

```python
w_types = soup.find_all('span', attrs={"class": "css-o1vzmt eoyjyou0"})
work_type = [w_type.text for w_type in w_types]
work_type
```

```python
df = pd.DataFrame({"Job": names,
                   "Company": companies,
#                   "Experience": years_of_exp,
                  "Location": locations,
                   "DatePosted": date_posted,
                  "JobType": job_type,
                  "WorkType": work_type})
df
```
